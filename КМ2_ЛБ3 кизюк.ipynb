## Задание 3.4. Движение секущей прямой к заданной линии на плоскости (версия 2)


```python
A_start = np.array([x1_array[0], y1_array[0]])  # начальное положение A (движется)
B_fixed = np.array([x1_array[-1], y1_array[-1]])  # неподвижная точка B

# Вычисление касательной в точке B
t_sym = sp.Symbol('t')
x_sym = sp.log(t_sym + sp.sqrt(t_sym**2 + 1))
y_sym = t_sym * sp.sqrt(t_sym**2 + 1)

dx_dt = sp.diff(x_sym, t_sym)
dy_dt = sp.diff(y_sym, t_sym)

# Производные в точке t_max
dx_at_B = dx_dt.subs(t_sym, t_max)
dy_at_B = dy_dt.subs(t_sym, t_max)

tangent_dir_B = np.array([float(dx_at_B), float(dy_at_B)])

# Параметры для отрисовки касательной в точке B
t_range = np.linspace(-2, 2, 100)
tangent_x_B = B_fixed[0] + tangent_dir_B[0] * t_range
tangent_y_B = B_fixed[1] + tangent_dir_B[1] * t_range

# СОЗДАЁМ НОВУЮ ФИГУРУ
fig8 = plt.figure(figsize=(8, 6))
ax8 = plt.axes()
plt.axis([x1_min-0.1, x1_max+0.1, y1_min-0.1, y1_max+0.1])

# Кривая
curve8, = ax8.plot(x1_array, y1_array, 'b', linewidth=2, label='Кривая')

# Неподвижная точка B (красная)
point_B8, = ax8.plot(B_fixed[0], B_fixed[1], 'ro', markersize=8, label='Точка B (неподвижная)')

# Подвижная точка A (зелёная)
point_A8, = ax8.plot(A_start[0], A_start[1], 'go', markersize=8, label='Точка A (движется)')

# Секущая (через A и B)
secant_line8, = ax8.plot([], [], 'g', linewidth=2, label='Секущая')

# Касательная в точке B (красная, пока невидимая)
tangent_line8, = ax8.plot(tangent_x_B, tangent_y_B, 'r--', linewidth=2, alpha=0, label='Касательная в B')

plt.legend()
plt.grid(True)

def init8():
    point_A8.set_data([A_start[0]], [A_start[1]])
    secant_line8.set_data([], [])
    tangent_line8.set_alpha(0)
    return point_A8, secant_line8, tangent_line8

def at_frame8(point):
    A = np.array([point[0], point[1]])  # текущее положение A
    
    point_A8.set_data([A[0]], [A[1]])
    secant_line8.set_data([A[0], B_fixed[0]], [A[1], B_fixed[1]])
    
    # Если A приблизилась к B, показываем касательную в B
    if np.linalg.norm(A - B_fixed) < 1e-6:
        tangent_line8.set_alpha(1)
    else:
        tangent_line8.set_alpha(0)
    
    return point_A8, secant_line8, tangent_line8

# Анимация (A движется от начала к концу)
ani8 = FuncAnimation(fig8, at_frame8, frames=matrix1, 
                     init_func=init8, repeat=False, interval=20, blit=True)
plt.show()
```
