## Задание 7.1 Определение классов

**Бинарное дерево** — это корневое дерево, вершины которого, исключая корень,содержатся в двух непересекающихся множествах: левое поддерево и правоеподдерево. При этом каждое поддерево либо является пустым множеством, либо бинарным деревом  

**Бинарное дерево поиска (binary search tree)** — это бинарное дерево, каждая вершина которого имеет значения одного типа, для которых определены операции сравнения например, значениями вершин являются объекты типа int или float). При этом значения вершин непустого левого поддерева меньше (<), чем значение соответствующей родительской вершины, а значения вершин непустогоправого поддерева больше либо равны (>=) значению соответствующей родительской вершины.  

**Бинарное дерево поиска** — это упорядоченная динамическая структура данных,которая позволяет быстро производить поиск элементов, а также быстро удалять и добавлять элементы.

**Рекурсивное определение:бинарное дерево** – это структура данных,которая либо пуста, либо содержит элемент дерева (корневую вершину), связанныйс двумя различными бинарными деревьями, называемыми левым и правым поддеревьями.

**Класс BinaryTree** является основным. Он определяет шаблон для создания объекта, представляющего бинарное дерево поиска. 

**Класс BinaryNode** является вспомогательным. Он определяет шаблон для создания объекта, представляющего элемент (вершину) бинарного дерева поиска.  

**Класс EmptyNode** является вспомогательным. Он определяет шаблон для создания объекта, представляющего "пустой" элемент (вершину) бинарного дерева поиска без значения и без поддеревьев. Экземпляры класса EmptyNode будут использоваться для представления пустого дерева и пустых поддеревьев.

````python
class BinaryTree:
    def __init__(self) -> None:
        self.root = EmptyNode()  # Изначально дерево пустое (корень - пустой узел)
        BinaryNode.numberOfNodes = 0  # Сброс счетчика узлов
    
    def __repr__(self) -> str:
        return repr(self.root)  # Возвращает строковое представление дерева
    
    def insert(self, value) -> None:
        self.root = self.root.insert(value)  # Вставка значения в дерево
    
    def __contains__(self, value) -> bool:
        return value in self.root  # Проверка наличия значения в дереве
    
    def __len__(self) -> int:
        return len(self.root)  # Возвращает количество узлов в дереве
````

````python
class BinaryNode:
    numberOfNodes = 0  # Статическая переменная - счетчик всех узлов
    
    def __init__(self, left, value, right) -> None:
        self.left = left    # Левый потомок
        self.value = value  # Значение узла
        self.right = right  # Правый потомок
        BinaryNode.numberOfNodes += 1  # Увеличиваем счетчик узлов
    
    def __repr__(self) -> str:
        # Строковое представление узла в формате (left,value,right)
        return f'({self.left},{self.value},{self.right})'
    
    def insert(self, value):
        # Рекурсивная вставка значения
        if value < self.value:
            self.left = self.left.insert(value)  # Вставка в левое поддерево
        else:
            self.right = self.right.insert(value)  # Вставка в правое поддерево
        return self
    
    def __contains__(self, value) -> bool:
        # Рекурсивный поиск значения
        if value == self.value:
            return True
        elif value < self.value:
            return value in self.left  # Поиск в левом поддереве
        else:
            return value in self.right  # Поиск в правом поддереве
    
    def __len__(self) -> int:
        # Возвращает общее количество узлов (использует статическую переменную)
        return BinaryNode.numberOfNodes
````

````python
class EmptyNode:
    def __repr__(self) -> str:
        return '*'  # Представление пустого узла
    
    def insert(self, value):
        # Создает новый узел при вставке в пустое место
        return BinaryNode(self, value, self)
    
    def __contains__(self, value) -> bool:
        return False  # В пустом узле ничего нет
    
    def __len__(self) -> int:
        return 0  # Пустой узел не учитывается в длине
````

````python
# 1: Создаем пустое дерево
tree = BinaryTree()
print("Пустое дерево:", tree)
print("Длина дерева (количество узлов):", len(tree))
print("Содержит 5?", 5 in tree)
print()

# 2: Вставляем элементы
tree.insert(10)
print("После вставки 10:", tree)
print("Длина:", len(tree))
print()

# 3: Вставляем еще элементы
tree.insert(5)
tree.insert(15)
tree.insert(3)
tree.insert(7)
tree.insert(12)
tree.insert(18)

print("После вставки 5,15,3,7,12,18:", tree)
print("Длина:", len(tree))
print()

# 4: Проверяем наличие элементов
print("Проверка наличия элементов:")
print("10 в дереве?", 10 in tree)
print("5 в дереве?", 5 in tree)
print("20 в дереве?", 20 in tree)
print("3 в дереве?", 3 in tree)
````

## Задание 7.2 Метод вставки элемента в бинарное дерево поиска


````python
class BinaryNode:
    numberOfNodes = 0  # Статическая переменная - счетчик всех узлов
    
    def __init__(self, left, value, right):
        self.left = left    # Левый потомок
        self.value = value  # Значение узла
        self.right = right  # Правый потомок
        BinaryNode.numberOfNodes += 1
    
    def __repr__(self):
        return f'({self.left},{self.value},{self.right})'
    
    def insert(self, value):
        """
        Метод вставки значения в бинарное дерево поиска
        """
        if value < self.value:
            self.left = self.left.insert(value)
        elif value > self.value:
            self.right = self.right.insert(value)
        # Если value == self.value - ничего не делаем (дубликаты не храним)
        return self
    
    def __contains__(self, value):
        """Проверка наличия значения в дереве"""
        if value == self.value:
            return True
        elif value < self.value:
            return value in self.left
        else:
            return value in self.right
    
    def __len__(self):
        """Возвращает общее количество узлов в дереве"""
        return BinaryNode.numberOfNodes
````

````python
print("Пример 1: Итерируемый объект - строка")

# Создаем дерево
tree1 = BinaryTree()

# Вставляем символы строки
for char in "программирование":
    tree1.insert(char)

print(f"Дерево: {tree1}")
print(f"Количество узлов: {len(tree1)}")
print(f"Уникальные символы: {len(set('программирование'))}")
print(f"Проверка: буква 'п' есть? {'п' in tree1}")
print(f"Проверка: буква 'ф' есть? {'ф' in tree1}")
````

## Задание 7.4 Перегрузка операции принадлежности in

````python
class EmptyNode:
    def __repr__(self):
        return '*'
    
    def insert(self, value):
        return BinaryNode(self, value, self)
    
    def __contains__(self, value):
        """Пустой узел никогда не содержит значений"""
        return False
    
    def __len__(self):
        return 0


class BinaryNode:
    numberOfNodes = 0
    
    def __init__(self, left, value, right):
        self.left = left
        self.value = value
        self.right = right
        BinaryNode.numberOfNodes += 1
    
    def __repr__(self):
        return f'({self.left},{self.value},{self.right})'
    
    def insert(self, value):
        if value < self.value:
            self.left = self.left.insert(value)
        elif value > self.value:
            self.right = self.right.insert(value)
        return self
    
    def __contains__(self, value):
        """
        Рекурсивная проверка наличия значения в дереве.
        Возвращает True, если значение найдено, иначе False.
        """
        if value == self.value:
            return True
        elif value < self.value:
            return value in self.left  # Рекурсивный поиск в левом поддереве
        else:
            return value in self.right  # Рекурсивный поиск в правом поддереве
    
    def __len__(self):
        return BinaryNode.numberOfNodes


class BinaryTree:
    def __init__(self):
        self.root = EmptyNode()
        BinaryNode.numberOfNodes = 0
    
    def __repr__(self):
        return repr(self.root)
    
    def insert(self, value):
        self.root = self.root.insert(value)
    
    def __contains__(self, value):
        """
        Перегрузка оператора in для дерева.
        Проверяет, содержится ли значение в дереве.
        """
        return value in self.root
    
    def __len__(self):
        return len(self.root)
````

## Задание 7.5 Перегрузка встроенной функции len

````python
class EmptyNode:
    def __repr__(self):
        return '*'
    
    def insert(self, value):
        return BinaryNode(self, value, self)
    
    def __contains__(self, value):
        return False
    
    def __len__(self):
        """
        Возвращает количество узлов в пустом поддереве.
        Пустой узел не содержит элементов, поэтому возвращает 0.
        """
        return 0


class BinaryNode:
    numberOfNodes = 0  # Статический счетчик всех узлов
    
    def __init__(self, left, value, right):
        self.left = left
        self.value = value
        self.right = right
        BinaryNode.numberOfNodes += 1
    
    def __repr__(self):
        return f'({self.left},{self.value},{self.right})'
    
    def insert(self, value):
        if value < self.value:
            self.left = self.left.insert(value)
        elif value > self.value:
            self.right = self.right.insert(value)
        return self
    
    def __contains__(self, value):
        if value == self.value:
            return True
        elif value < self.value:
            return value in self.left
        else:
            return value in self.right
    
    def __len__(self):
        """
        Возвращает общее количество узлов в дереве.
        Использует статическую переменную numberOfNodes.
        """
        return BinaryNode.numberOfNodes


class BinaryTree:
    def __init__(self):
        self.root = EmptyNode()
        BinaryNode.numberOfNodes = 0  # Сбрасываем счетчик при создании нового дерева
    
    def __repr__(self):
        return repr(self.root)
    
    def insert(self, value):
        self.root = self.root.insert(value)
    
    def __contains__(self, value):
        return value in self.root
    
    def __len__(self):
        """
        Перегрузка встроенной функции len().
        Возвращает количество вершин в бинарном дереве поиска.
        """
        return len(self.root)
````

## Задание 7.6 Трассировка объектного уровня

### Задание 1. Вставка новой вершины в непустое бинарное дерево

**Процесс вставки вершины в непустое дерево:**

1. **Вызов метода `insert` у дерева:**
   - При вызове `tree.insert(value)` сначала вызывается метод `insert` класса `BinaryTree`.
   - Этот метод делегирует вставку корневому узлу: `self.root = self.root.insert(value)`.

2. **Обработка вставки в `BinaryNode`:**
   - Если дерево непустое, `self.root` — это объект класса `BinaryNode`.
   - Метод `insert` класса `BinaryNode` сравнивает `value` с текущим значением узла (`self.value`):
     - Если `value < self.value`, рекурсивно вызывается `self.left.insert(value)`.
     - Если `value >= self.value`, рекурсивно вызывается `self.right.insert(value)`.
   - Рекурсия продолжается до достижения `EmptyNode`.

3. **Достижение `EmptyNode`:**
   - Когда рекурсивный вызов доходит до `EmptyNode`, его метод `insert` создает новый `BinaryNode` с переданным значением.
   - Новый узел инициализируется с `left` и `right`, ссылающимися на `EmptyNode`.

4. **Обновление ссылок:**
   - Ссылка на поддерево (`left` или `right`) родительского узла обновляется, чтобы указывать на новый узел.

**Пример для `tree.insert(1)` (из `source_data = [5, 1, 10, 3, 4]`):**
- Сравниваем `1` с корнем `5`: `1 < 5` → идем влево.
- Левый узел корня — `EmptyNode`. Создаем новый `BinaryNode` со значением `1`.
- Обновляем `left` корневого узла, чтобы он ссылался на новый узел.


### Задание 3. Трассировка вставки `tree.insert(10)`

**Исходное состояние дерева после вставки 5 и 1:**
```
      5
     /
    1
```

**Шаги вставки `10`:**
1. **Вызов `tree.insert(10)`:**
   - Метод `insert` класса `BinaryTree` вызывает `self.root.insert(10)`.

2. **Сравнение в корневом узле (`5`):**
   - `10 > 5` → идем в правое поддерево (`self.right`).
   - Правое поддерево корня — `EmptyNode`.

3. **Создание нового узла:**
   - Метод `insert` класса `EmptyNode` создает `BinaryNode(EmptyNode(), 10, EmptyNode())`.
   - Правая ссылка корневого узла обновляется на новый узел.

**Итоговая структура:**
```
      5
     / \
    1   10
```

**Объекты в памяти:**
- `tree` (BinaryTree) → `root` (BinaryNode, значение `5`):
  - `left` (BinaryNode, значение `1`):
    - `left` и `right` ссылаются на `EmptyNode`.
  - `right` (BinaryNode, значение `10`):
    - `left` и `right` ссылаются на `EmptyNode`.
