### 2.1 Понятие объекта. Атрибуты. Методы. Встроенные типы объектов

**Объект** в Python — это базовая сущность для представления данных.

К объектам в Python относятся:
- данные встроенных типов
- встроенные и пользовательские функции
-  встроенные и пользовательские классы
-   экземпляры
-   классов
-   модули
  
Внутренне объекты представляют собой области выделенной памяти с достаточным
пространством для хранения информации об объекте. Информация об объекте
делится на данные объекта (значение объекта, состояние объекта) и метаданные.


**Метаданные** объекта называются атрибутами. **Атрибут** — это имя переменной или
имя функции, которое связано с объектом и используется для получения
информации об объекте и совершения действий над этим объектом. Все атрибуты
объекта составляют пространство имен объекта. Атрибуты, которые начинаются и заканчиваются двумя знаками подчеркивания,
чаще всего являются **методами**, реализующими операции над объектами.

**Операция (operator)** — это конструкция языка, аналогичная по записи
математическим операциям. Для обозначения операции часто используют не
алфавитно-цифровые, а специальные символы.


**Встроенные типы объектов делятся на числа и коллекции.**
    
Типы объектов, представляющих числа: целые числа ( int ), вещественные числа
( float ), комплексные числа ( complex ). Булевый тип ( bool ) наследуется от типа int , поэтому булевый тип относится к типам числовых объектов.

Коллекции — это наборы объектов, которые делятся на три вида:
- Последовательности — это позиционно упорядоченные коллекции. Встроенные
типы, относящиеся к последовательностям: строки (str), списки ( list ),
кортежи ( tuple ).
- Отображения — это неупорядоченные коллекции объектов, хранящихся по
ключам. Встроенные типы, относящиеся к отображениям: словарь ( dict ).
- Множества — это неупорядоченные коллекции не повторяющихся объектов.
Встроенные типы, относящиеся к множествам: множество ( set ).

**Литерал** в Python — это форма записи данных для их представления объектом
определеннного типа. 

**Выражение (expression)** в Python — это конструкция языка в виде комбинации
литералов, переменных, операций и вызовов функций, которая вычисляется и
возвращает значение в виде объекта. Выражение не хранится в памяти, в памяти
сохраняется результирующий объект.

**1. Метод .join()**

- Синтаксис: строка_разделитель.join(итерируемый_объект)
- Назначение: Объединяет элементы итерируемого объекта (списка, кортежа и т.д.) в одну строку.
- Параметры:
  · итерируемый_объект: обязательный параметр. Это коллекция (например, список), содержащая строки, которые нужно соединить. Важно: все элементы в коллекции должны быть строками. Если там будет число, Python выдаст ошибку.
- Возвращаемое значение: Новая строка, состоящая из элементов переданной коллекции, разделенных строкой-разделителем.
- Особенности: Это метод строки, а не списка. Вызывается он именно от той строки, которая будет служить разделителем.


**2. Метод .split()**

- Синтаксис: исходная_строка.split(разделитель, максимальное_количество_разбиений)
- Назначение: Разбивает исходную строку на части (подстроки) по указанному разделителю.
- Параметры:
  · разделитель: необязательный параметр. Строка, которая будет считаться границей между частями. Если его не указать (или указать None), метод будет разбивать строку по последовательностям пробельных символов (пробелы, символы табуляции, перевода строки), игнорируя их в начале и конце строки.
  · максимальное_количество_разбиений: необязательный параметр. Целое число, указывающее, сколько разбиений выполнить максимум. Оставшаяся часть строки будет последним элементом списка.
- Возвращаемое значение: Список (list) строк, полученных после разбиения.
- Особенности: Это метод строки. Он всегда возвращает список.

### 2.9 Переменная


**Переменная (variable)** - это набор символов (имя переменной) для хранения адреса
памяти, где находится объект.

еременная создается при первом присваивании ей значения. Основной способ
создания переменных – это использование оператора присваивания = :
<переменная> = <значение переменной> . При этом значением переменной
является объект произвольного типа.

Понятие типа не связано с переменной, а связано со значением переменной. В
процессе выполнения кода значение переменной можно изменять за счет
повторных присваиваний одной и той же переменной объектов разных типов.
Такая возможность Python называется **динамической типизации переменных**.


Переменная в Python является ссылкой на объект в памяти. Внутренне ссылка
записывается в системной таблице в виде соответствия между именем переменной
и адресом ячейки памяти, где располагается объект. Переменная не хранит объект.
Переменная в каждый момент выполнения кода ссылается на один определенный
объект.


Если объект коллекции содержит ссылку на самого себя, то он называется
**циклическим объектом**. Когда обнаруживается цикл в объекте, то он выводится
специальном образом без возникновения зацикливания

```python
a = [4, 9]; a.append(a)
a
```

```python
d={'x':1}
d['self']=d
d
```

```python
t=(1,2,[])
t[2].append(t)
t
```

Если несколько переменных ссылаются на один и тот же объект, то этот объект
называется **разделяемым объектом**, а ссылка на один и тот же объект называется
**разделяемой ссылкой**

```python
a = b = c = d = [4]
id(a), id(b), id(c)
```

В Python поддерживается особый объект-заполнитель None , часто используемый
для инициализации переменных в случае, когда значение переменной будет
определено позже.

Когда утрачивается последняя ссылка на объект, например, за счет присваивания
единственной переменной, ссылающейся на объект, чего-нибудь другого, то вся
область памяти, занимаемая исходным объектов, очищается автоматически
(сборка мусора). **Автоматическое управление памятью** с помощью Python
Memory Manager является спецификой языка Python.


Для того, чтобы различные переменные не ссылались на один и тот же объект,
нужно создавать копии объектов. Существует три способа создания копий
объектов.

**Способ 1 (универсальный)**. Копирование объектов возможно с помощью вызова
соответствующих функций-конструкторов.
Создаем копию исходного объекта с помощью функции конструкторв; убеждаемся
в том, что уникальные идентификаторы исходного и построенного объектов не
совпадают

```python
obj = [1, 2, 3]; print(obj, id(obj))
obj_copy = list(obj); obj_copy = obj.__class__(obj) # два способа
print(obj_copy, id(obj_copy))
```

**Способ 2** (для последовательностей). Копирование последовательностей ( str ,
list , tuple ) возможно посредством операции среза по всем элементам [:]

```python
objs2 = 'Hellow', [1, 2, 3], (4,5,6)
objs2_copy = tuple(map(lambda x: x[:], objs2))
objs2_copy
```

**Способ 3** (для изменяемых объектов). Копирование изменяемых объектов (списков
( list ), словарей ( dict ) и множеств ( set )) возможно с помощью метода copy

```python
obj3 = [[1],2,3];
c = obj3.copy(); c[0][0] = 0; c[1] = 20; obj3, c
```

```python
LIST = [1, 2, 3]
a = LIST[:]
b = list(LIST)
c = LIST.copy()
print(a, b, c)
```

### 2.10 Выражение спискового включения

**Списковое включение** (list comprehension) это выражение, которое создает
НОВЫЙ СПИСОК за счет выполнения выражения на каждом элементе
последовательности (итерируемого объекта), по одному за раз.
Выражение спискового включения записывается в квадратных скобках и содержит
конструкцию цикла for ... in ... .


Целесообразно применять списковое включение тогда, когда необходимо
выполнить выражение на каждом элементе итерируемого объекта и получить
список в результате.


Синтаксис списковых включений позволяет дополнительно задавать ограничения
на элементы из последовательности (итерируемого объекта) с использованием
конструкции выбора if , на которых будет выполняться выражение.

```python
[x**2 for x in range(10) if x % 2 == 0]
```

```python
a = {i for i in range(1,20) if i % 3 ==0}
a
```

```python
a = [[1 for i in range(8)] for j in range(7)]
a
```

```python
import numpy as np
matrix = np.ones((7, 8))
print(matrix)
```

```python
import numpy as np
matrix = np.ones((7, 8),dtype=int)
print(matrix)
```

```python
a = {i : {x : s for (x,s) in enumerate('hellow')} for i in range(1, 7)}
a
```

### 2.11 Форматированный вывод строк

Существует три способа форматированного вывода строк на основе подстановок в
исходную форматную строку объектов произвольных типов согласно шаблону:
- бинарная операция % : '...%s...%d...' % (obj1, obj2)
- строковый метод format : '...{}...{}...'.format(obj1, obj2)
- форматированный строковый литерал f'...{x}...{y}...'

```python
name = "Анастасия"
age = 17

# Синтаксис: f"текст {переменная} текст"
result = f"Меня зовут {name}, мне {age} лет"
print(result)
```

```python
# Синтаксис: "текст {}".format(переменная)
result1 = "Меня зовут {}, мне {} лет".format(name, age)
print(result1)
```

```python
# Синтаксис: "текст %s %d" % (переменная1, переменная2)
result2 = "Меня зовут %s, мне %d лет" % (name, age)
print(result2)
```

```python
a = 9
b = 5

print(f'expr = {a+b}')
print('expr = %s' %(a+b))
print('expr = {0}'.format(a+b))
```

### 2.12 Файловый объект

**Файловый объект** -- это объект, предоставляющий интерфейс для работы с файлами
на диске или устройствами хранения и передачи данных (стандартный поток ввода/
вывода, буфер памяти, сокет и др.).


Файловый объект в Python создается вызовом встроенной функции open(file,
mode) , аргументами которой являются имя файла file в виде строки и режим
обработки файла mode в виде строки.

- **Текстовые файлы** представляют содержимое в виде строк типа str ,
автоматически выполняя кодирование и декодирование Unicode при
записывании и чтении данных.
- **Двоичные файлы** представляют содержимое в виде специальных строк
bytes и позволяют получать доступ к содержимому файла, не изменяя его.


Встроенный тип данных **байтовая строка** (bytes) — это последовательность
символов для хранения и представления байтовой информации, используемой в
файлах медиа-данных и при передаче через сеть.


```python
import random

n=15
numbers=[random.randrange(100) for _ in range(n)]
         
f = open('tmp.txt','w')
f.write("".join([str(x)+'\n' for x in numbers]))
f.close
f=open('tmp.txt','r')
print(f.read())
f.close;
```

```python
lines = ["строка 1\n", "строка 2\n", "строка 3\n"]
f = open('test.txt', 'w')
f.writelines(lines)
f.close()

f = open('test.txt', 'r')
content = f.readlines()
f.close()

print(content)
```
