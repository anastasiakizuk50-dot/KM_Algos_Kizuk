## Этап 3. Определение функции перемещения и
## нахождение угла запуска тела


Полагаем, что начало прямоугольной декартовой системы координат соответствует
уровню земли. Тогда в начальный момент движения тела $t=0$ имеем, что  $$s_{x}(0) = 0,  s_{y}(0)= h_{start}$$


Для построения функции перемещения $s_{x}(t),s_{y}(t)$ воспользуемся вторым
законом Ньютона $$m\bf a=\bf F,$$

где $m$-масса тела,ускорение движения тела а = $(s"_{x}(t), s"_{y}(t))$ и приложенная сила ${\bf F} = m (0,-g).$ В результате имеем два уровнения, что $$s"_{x}(t) = 0,  s"_{y}(t)= -g $$

На основании уравнений с учетом начальных условий можно записать
аналитический вид для функций  $s_{x}(t)$ и $s_{y}(t)$ $$s_{x}(t)=s_{x}(0)+v_{0x}(t)*t=0+v_{0x}(t)*t=v_{0x}(t)*t$$ $$s_{y}(t)=s_{y}(0)+v_{0y}(t)*t-\frac{g^2}{2}t^2= h_{start}+v_{0y}(t)*t-\frac{g^2}{2}t^2$$

где $v{0}=(v_{0x},v_{0y})$обозначает неизвестную скорость движения тела в начальный момент времени t=0.

<span style='color:red'>Напишите</span> подробно, как из второго закона Ньютона и начальных условий получен аналитический вид функции перемещения $s_x(t), s_y(t)$. Объяснения оформите в тексте документа с лабораторной работой.

Движение тела является равноускоренным, так как на него действует постоянная сила тяжести. Из школьной программы известно, что для равноускоренного движения координата тела изменяется по закону:

$$x = x_0 + v_0 t + \frac{a t^2}{2}$$

В нашем случае из второго закона Ньютона $m\vec{a} = \vec{F}$ и силы тяжести $\vec{F} = m\vec{g}$ получаем, что ускорение $\vec{a} = \vec{g} = (0, -g)$.

Рассмотрим движение по отдельности по каждой оси.

По оси $Ox$ ускорение отсутствует ($a_x = 0$), следовательно, движение является равномерным. Тогда из общей формулы при $a_x = 0$ и $x_0 = 0$ (тело начинает движение из начала координат по горизонтали) получаем:

$$s_x(t) = v_{0x} t$$

По оси $Oy$ ускорение постоянно и равно $-g$. Движение является равноускоренным (точнее, равнозамедленным, так как ускорение направлено вниз). Подставляем $a_y = -g$, начальную координату $y_0 = h_{\text{start}}$ и начальную скорость $v_{0y}$:

$$s_y(t) = h_{\text{start}} + v_{0y} t + \frac{(-g) t^2}{2}$$

Отсюда окончательно получаем:

$$s_y(t) = h_{\text{start}} + v_{0y} t - \frac{g}{2} t^2$$

Таким образом, функции перемещения имеют вид:

$$\boxed{s_x(t) = v_{0x} t}$$
$$\boxed{s_y(t) = h_{\text{start}} + v_{0y} t - \frac{g}{2} t^2}$$


$\color{red}\text{Задание* необязательное}$ как изменится вид функций перемещения,если начальный момент времени будет равен не t=0, а t=t*?

 $$s_{x}(t)= s_{x}(t^*)+ v_{x}(t^*)*(t-t^*),$$
 $$s_{y}(t)=s_{y}(t^*)+ v(t^*)_{y}(t-t^*)-\frac{g}{2}*(t-t^*)^2,$$

Из условия задачи имеем, что $s_{x}(T)=s_{end}$ Из этого соотношения находим значение для $v_{0x}$

```python
v0_x = s_end/T
v0_x
```

Из условия задачи имеем, что $s_{y}(T)=h_{end}$ Из этого соотношения находим значение для $v_{0y}$

```python
v0_y = (h_end-h_start+g/2*T**2)/T
v0_y
```

Значение угла запуска тела $\alpha$ определяем из соотношения tan  $\alpha=\frac{v0_{y}}{ v0_{x}}$

```python
alpha = math.atan(v0_y/v0_x)
alpha, math.degrees(alpha) # функция degrees преобразует угол из радиан в градусы
```

Результирующая функция перемещения имеет следующий вид:

# Задание 1.2. Определение момента
# времени касания телом земли

Выполнение задания осуществите на основе двух подходов.

*Подход 1* (**точное решение**): время касания телом земли находим как точное
решение уравнения $s_{y}(t)=0$

*Подход 2* (**приближенное решение**)
- Постройте последовательность $s_{y}$ на временном отрезке [0,$t_{stop}$] , чтобы момент касания тела земли $t^*$ попадал в выбранный отрезок, т.е.    t* $\in$ [0,$t_{stop}$].Отметим, что $t_{stop}>T$, где значение $T$ задается в Задании 1.1.
- Найдите в построенной последовательности $s_{y}$ два соседних элемента последовательности, значения которых меняют знак. Т. е. найдите значение индекса k, для которого $s_{y}[k]s_{y}[k+1]<0$ . Это будет означать, что 0 $\in$ ($s_{y}[k],s_{y}[k+1]).$
- Постройте прямую линию $s_{lin}(t)$, проходящую через две точки $(t_{k},s_{y}[k])$ и  $(t_{k+1},s_{y}[k+1])$ и найдите значение t* $\in$  $(t_{k},t_{k+1})$  из условия $s_{lin}(t*)=0$. Найденное значение t* , будет соответствовать точке (t*,0) на
прямой $s_{lin}(t)$. $\color{red}\text{Приведите}$ в документе формулы для прямой $s_{lin}(t)$ и для вычисления t* .
$\color{red}\text{Сравните}$ точное решение (Подход 1) и приближенное решение
(Подход 2), вычислив относительную ошибку.

**Формула прямой $s_{lin}(t):$** 
Здесь мы используем линейную интерполяцию. Мы не знаем точного закона движения между двумя точками $k$ и $k+1$, но мы предполагаем, что на этом маленьком отрезке функцию $s_{y}(t)$ можно приблизить прямой линией. Коэффициент $\frac{s_y[k+1] - s_y[k]}{t_{k+1} - t_k}$ — это скорость изменения координаты (приближенная вертикальная скорость) на этом участке.
$$s_{lin}(t)=s_{y}[k]+\frac{s_y[k+1] - s_y[k]}{t_{k+1} - t_k}*(t-t_{k})$$
**Формула $t^*$:** Из условия:$s_{lin}(t^*)=0$, имеем: $$0=s_{y}[k]+\frac{s_y[k+1] - s_y[k]}{t_{k+1} - t_k}*(t^*-t_{k})$$ выражаем $t^*$ :
$$t^* = \frac{-s_{y}[k]*(t_{k+1}-t_{k})}{s_{y}[k+1]-s_{y}[k]}+t_{k}$$

Относительная ошибка двух подходов зависит от выбора шага интерполяции. В нашем случае ошибка составляет приблизительно $0.0002$%
